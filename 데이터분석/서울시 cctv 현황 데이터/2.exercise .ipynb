```python
cctv.rename(columns={'기관명' : '자치구'}, inplace=True)  # 
cctv.head()

#pop_Seoul.rename(columns={pop_Seoul.columns[0] : '구별',
#                          pop_Seoul.columns[1] : '인구수',
#                          pop_Seoul.columns[2] : '한국인',
#                          pop_Seoul.columns[3] : '외국인',
#                          pop_Seoul.columns[4] : '고령자'}, inplace=True)
#pop_Seoul.head()
```

```python
#cctv[['2011년 이전','2012년','2013년']] = \
#    cctv[['2011년 이전','2012년','2013년']].fillna(0).astype(int) 
#cctv
cctv['2011년 이전'] = cctv['2011년 이전'].fillna(0).astype(int)
cctv['2012년'] = cctv['2012년'].fillna(0).astype(int)
cctv['2013년'] = cctv['2013년'].fillna(0).astype(int)
cctv
```

```python
#cctv['2015년 이전'] = cctv[['2011년 이전','2012년','2013년','2014년','2015년']].sum(axis=1)
cctv['2015년 이전'] = cctv.iloc[:,2:7].sum(axis=1)
cctv.head()
```

```python
# 최근 3개년간의 데이터를 합하여 '최근 3개년' 컬럼을 만든 후, '
# 최근 증가율' 컬럼을 만든다.
#cctv['최근 3개년'] = cctv[['2016년','2017년','2018년']].sum(axis=1)
cctv['최근 3개년'] = cctv.iloc[:,7:10].sum(axis=1)
cctv.head()

```

```python
cctv['최근 증가율'] = (cctv['최근 3개년']-cctv['2015년 이전'])/cctv['2015년 이전']*100
cctv.head()
```

```python
# 자치구', '소계', '최근 증가율' 컬럼을 제외하고 나머지는 모두 삭제한다.
#cctv.drop['2011년 이전','2012년','2013년','2014년','2015년'], axis=1, inplace=True
#del cctv['최근 3개년']
#del cctv['2018년']
#del cctv['2017년']
#del cctv['2016년']
#del cctv['2015년']
#del cctv['2014년']
#del cctv['2013년']
#del cctv['2012년']
#del cctv['2011년 이전']
cctv.drop(cctv.iloc[:,2:11], axis=1, inplace=True)
cctv.head()
```

```python
# 서울시 인구 데이터와 cctv 데이터를 합치고 자치구를 인덱스로 만든다.
# 두 데이터를 합 할때 join 

df = pd.merge(cctv,pop, on='자치구')   # 겹치는 부분을 on = ''
df.head()
```

```python
df.set_index('자치구', inplace =True)
df.head()
```

```python
# 자치구 별 cctv 갯수
# sort_values() 내림차순
df['소계'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
# 자치구 별 인당 CCTV 갯수
df['인당 CCTV'] = df['소계'] / df['인구수'] 
df['인당 CCTV'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))

plt.show()
```

```python
df['최근 증가율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

#### = 인구수와 CCTV 댓수와 산점도

```python
plt.figure(figsize=(10,10))
plt.scatter(df['인구수'], df['소계'], s=50)  # s=50 점크기
plt.xlabel('인구수')
plt.ylabel('cctv 대수')
plt.grid()
plt.show()
```

#### 인구수와 CCTV 댓수와 산점도 및 희귀선

```python
fp1 = np.polyfit(df['인구수'], df['소계'], 1)
fp1   # 결과값 x 기울기 , y 절평
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 101)  # linspace 1차원 배열 만들기
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df['인구수'], df['소계'], s=50)  # s=50 점크기
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='m')
plt.xlabel('인구수')
plt.ylabel('cctv 대수')
plt.grid()
plt.show()
```

#### 인구수와 CCTV 댓수의 산점도, 회귀선 및 자치구명

```python
df.head()
```

```python
df['오차'] = np.abs(df['소계'] - f1(df['인구수']))   # 열계산하여 생성하기
df.head()
```

```python
df_sort = df.sort_values(by='오차', ascending=False)  # ascending 오름차순 True 오름, False 내림차순
df_sort.head(10)   # head 기본적으로 5개 () 안에 숫자 기입하면 그 숫자 행이 출력됨
```

```python
plt.figure(figsize=(14,10))
plt.scatter(df['인구수'], df['소계'],
            c=df['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98,
            df_sort.index[n], fontsize=15)
   
plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()    # 오른쪽 컬러바 생성
plt.grid()        # 그래프 상에 선 생성
plt.show()
```
