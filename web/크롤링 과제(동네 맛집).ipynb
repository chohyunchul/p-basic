```python
name = []
category = []
location = []
number = []
page = []
rest_list = soup.find_all(class_='s_norm')

for rest in rest_list:
    name.append(rest.find(class_='shop_title').get_text())
    category.append(rest.find(class_='shop_category').get_text())
#     location.append(' '.join(rest.get_text().split()[5:9]))
#     number.append(rest.get_text().split()[-2])
    page.append(rest.find(class_='shop_title')['href'])
```

```python
import pandas as pd
df = pd.DataFrame({'이름':name, '구분':category, '홈페이지':page})
df
```

```python
number = []

for n in tqdm_notebook(df.index):
    html = urlopen(df['홈페이지'][n])
    soup_tmp = BeautifulSoup(html, 'html.parser')
#     print('성공')
    number.append(soup_tmp.find(class_='mz_tel').get_text().strip())
```

```python
df['전화'] = number 
```

```python
df = df[['이름','구분','전화','홈페이지']]
```
