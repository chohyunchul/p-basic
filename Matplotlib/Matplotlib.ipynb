```python
# 12-4 선 그래프의 범례
xdata =['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
plt.plot(xdata, ydata1, label='국어')
plt.plot(xdata, ydata2, label='영어')
plt.legend(loc='upper center')
plt.title('세명 학생의 국어, 영어 성적')

plt.show()
```

```python
# 12-5 그래프 선 스타일 설정
xdata =['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
ydata3 = [85,97,82]
ydata4 = [92,88,82]

plt.plot(xdata, ydata1, label='국어', color='red', linestyle='-', marker='o')
plt.plot(xdata, ydata2, label='영어', color='#00ffff', linestyle='--', marker='x')
plt.plot(xdata, ydata3, label='수학', color='magenta', linestyle='-.', marker='s')
plt.plot(xdata, ydata4, label='사회', color='#444444', linestyle=':', marker='d')

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc = 'best')

plt.show()
```

```python
# 12-5 그래프 선 스타일 설정 / 생략 버전
xdata =['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
ydata3 = [85,97,82]
ydata4 = [92,88,82]

plt.plot(xdata, ydata1, 'ro-', label='국어', linewidth=3, markersize=10)
plt.plot(xdata, ydata2, 'cx--',label='영어', linewidth=3, markersize=10)
plt.plot(xdata, ydata3, 'ms-.',label='수학', linewidth=3, markersize=10)
plt.plot(xdata, ydata4, 'kd:',label='사회', linewidth=3, markersize=10)

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc = 'best')

plt.show()
```

```python
# 12-6 x/y 축의 범위와 눈금 설정

xdata = list(range(40))
ydata= []
             
for x in xdata:
   y=2*x
   ydata.append(y)
             
plt.plot(xdata, ydata, label='y=2x')

plt.xlim(0,40)
plt.xticks(list(range(0,40,2)))

plt.ylim(0,80)
plt.yticks(list(range(0,80,5)))
             
plt.grid(True)
plt.show()
```

```python
# cos함수
import numpy as np                       # 배열생성을 위한 numpy모듈 호출
from matplotlib import pyplot as plt     # 그래프 출력을 위한 pyplot 모듈 호출

x = np.linspace(-np.pi,np.pi,300)        # x 좌표값 생성 (-pi ~ +pi)
y = np.cos(x)                            # y = cos(x) 계산

plt.plot(x,y)                            # 그래프에 x,y 입력
plt.show()   
```

```python
# sin cos 함수
import numpy as np
import matplotlib.pyplot as plt
 
# 데이터 준비
x = np.arange(0, 6, 0.1) # 0 에서 6까지 0.1 간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)
 
# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") # cos 함수는 점선으로 그리기
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title('sin &amp; cos') # 제목
plt.legend()
plt.show()
```
