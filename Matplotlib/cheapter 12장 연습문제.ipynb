```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
# 화면에 그래프 그리기
%matplotlib inline
# Korean Setting , Minus 표시
mpl.rcParams['axes.unicode_minus'] = False
# family = 폰트 이름 size = 사이즈 크기
mpl.rc('font', family='D2Coding', size=14)
# Graph 크기 변경
plt.rcParams["figure.figsize"]= (10,6)
font1 = {'size':18, 'color':'green'}
```

```python
#1
import csv

f=open('cafe_2year.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines)
print(header)

list_data=[]
list_aaaa=[]
for line in lines:
    if '12' in line[0] or '03' in line[0] or '06' in line[0] or '09' in line[0]:
        list_data.append(line[0])
        list_aaaa.append(line[1])   
        
plt.title('전국 영화관람 회수')
plt.bar(list_data,list_aaaa)
plt.xlabel('년월')
plt.ylabel('카페수')
plt.show()
```

```python
#2
import csv
doc =np.zeros((7, 4))

f = open('doctor_2019.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)

header = next(rdr)

# for line in rdr:
#    print(line)

 
for line in rdr:
        if line[0] =='서울':
            doc[0][0] +=int(line[2])
            doc[0][1] +=int(line[3])
            doc[0][2] +=int(line[4])
            doc[0][3] +=int(line[5])
            
        if line[0] =='부산':
            doc[1][0] +=int(line[2])
            doc[1][1] +=int(line[3])
            doc[1][2] +=int(line[4])
            doc[1][3] +=int(line[5])
        
        if line[0] =='대구':
            doc[2][0] +=int(line[2])
            doc[2][1] +=int(line[3])
            doc[2][2] +=int(line[4])
            doc[2][3] +=int(line[5])
            
        if line[0] =='인천':
            doc[3][0] +=int(line[2])
            doc[3][1] +=int(line[3])
            doc[3][2] +=int(line[4])
            doc[3][3] +=int(line[5])
            
        if line[0] =='대전':
            doc[4][0] +=int(line[2])
            doc[4][1] +=int(line[3])
            doc[4][2] +=int(line[4])
            doc[4][3] +=int(line[5])
            
        if line[0] =='광주':
            doc[5][0] +=int(line[2])
            doc[5][1] +=int(line[3])
            doc[5][2] +=int(line[4])
            doc[5][3] +=int(line[5])
            
        if line[0] =='울산':
            doc[6][0] +=int(line[2])
            doc[6][1] +=int(line[3])
            doc[6][2] +=int(line[4])
            doc[6][3] +=int(line[5])
doc = doc.astype('int32')
doc = doc.tolist()
doc
```

```python
#4
import csv

f = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f)
    
header = next(lines)

list_data=[]

for line in lines:
    list_data.append(line[:])

length = len(list_data)

xdata = []
ydata = []

for i in range(1,15,2):
    xdata.append(list_data[i][0])
    ydata.append(float(list_data[i][4]))

plt.scatter(xdata, ydata)
plt.ylabel('의사 수')
plt.xlabel('시')
plt.title('서울 및 광역시 종합병원 전문의 수', fontdict = font1 )
plt.show()

f.close()
```

```python
#5
import csv

f = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f)
    
header = next(lines)

xdata= []
ydata1= []
ydata2= []
ydata3= []
ydata4= []

list_data=[]

for line in lines:
    list_data.append(line[:])


length = len(list_data)

for i in range(1,15,2):
    xdata.append(list_data[i][0])
    ydata1.append(float(list_data[i][1]))
    ydata2.append(float(list_data[i][2]))
    ydata3.append(float(list_data[i][3]))
    ydata4.append(float(list_data[i][4]))

plt.plot(xdata, ydata1, 'ro-')
plt.plot(xdata, ydata2, 'bx--')
plt.plot(xdata, ydata3, 'ys:')
plt.plot(xdata, ydata4, 'md-.')
plt.title('서울 및 광역시 종합병원 의사 수', fontdict = font1 )
plt.show()

f.close()
```

```python
#6import csv

f = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f)
    
header = next(lines)

xdata= []
ydata1= []
ydata2= []
ydata3= []
ydata4= []

list_data=[]

for line in lines:
    list_data.append(line[:])

length = len(list_data)
portion = [list_data[1][1], list_data[1][2], list_data[1][3], list_data[1][4]]
docs = ['일반의', '인턴', '레지던트','전문의']

xdata.append(list_data[1][0])
ydata1.append(list_data[1][1])
ydata2.append(list_data[1][2])
ydata3.append(list_data[1][3])
ydata4.append(list_data[1][4])

plt.pie(portion, explode = (0,0.1,0,0), labels=docs, autopct='%1.1f%%', shadow = True, startangle =90) 
plt.title('서울 종합병원 의사 유형별 분포', fontdict = font1 )
plt.show()

f.close()
```
