```python
#13-3 pandas를 사용해여 dataframe 이용
df = pd.read_csv('high_school_2019.csv')
df
```

```python
df2 = df.iloc[:,2:]
df2 
```

```python
df2['총 학생수']=df2[' 1학년 학생수']+df2[' 2학년 학생수']+df2[' 3학년 학생수']
df2
```

```python
#13-4
a = df2[[' 1학년 학생수',' 2학년 학생수',' 3학년 학생수']].sum()
a
```

```python
a.plot.bar(rot=1)
plt.title('전국 고등학교 학년별 총 학생수')
plt.show()
```
